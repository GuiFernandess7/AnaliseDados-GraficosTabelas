## Remodelamento de dados

```
# Tabela de Produtos e Preços Unitários
precos = planilha[["Produto", "Preco Unitario"]]
precos.head()
```

```
# Faturamentos maiores de 10000 reais
maiores_de_10mil = planilha[planilha["Faturamento na Venda"] > 10000]

maiores_de_10mil
```

```
# Total de faturamento acima de 10000
total_linhas = planilha.shape
total_maiores = maiores_de_10mil.shape
print(f'Total de produtos com faturamente superior a 10000 reais: {total_maiores[0]}')
print(f'Total de linhas: {planilha.shape[0]}')
```

```
# Produtos
TVs = planilha[planilha["Produto"].isin(["Televisão"])]
Notebooks = planilha[planilha['Produto'].isin(['Notebook'])]
Iphones = planilha[planilha['Produto'].isin(['IPhone'])]
Tablets = planilha[planilha['Produto'].isin(['Tablet'])]
Androids = planilha[planilha['Produto'].isin(['Android'])]
SmartWatchs = planilha[planilha['Produto'].isin(['SmartWatch'])]
Cameras = planilha[planilha['Produto'].isin(['Câmera'])]

TVs
```

```
# Faturamento total de televisões 
TVs["Faturamento na Venda"].sum()
```
